```python
from bells_o import Evaluator, Usage, Result
from bells_o.evaluator import DatasetConfig, SupervisorConfig
from bells_o.supervisors.rest import LakeraGuardDefaultSupervisor
from bells_o import HuggingFaceDataset
from dotenv import load_dotenv

load_dotenv()
```

```python
def bells_target_map(input: str):
    return Result(content_moderation=1.0 if input == "harmful" else 0.0)


dataset_conf: DatasetConfig = DatasetConfig(
    type=HuggingFaceDataset,
    kwargs={
        "name": "brash6/content_moderation_bells_v2",
        "usage": Usage("content_moderation"),
        "target_map_fn": bells_target_map,
        "filters": {"harm_level": ["benign", "harmful"]},
    },
    input_column="question",
    target_column="harm_level",
)
```

```python
supervisor_conf = SupervisorConfig(
    type=LakeraGuardDefaultSupervisor,
    kwargs={"project_id": "project-4654781877", "api_variable": "LAKERA_API"},
)
```

```python
evaluator = Evaluator(
    dataset_conf, supervisor_conf, save_dir="../lakera_guard_default_results", verbose=True
)
```

```python
evaluator.run(run_id="run_1", verbose=True, save=True)
```
