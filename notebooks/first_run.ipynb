```python
from bells_o import Evaluator, Usage, Result
from bells_o.evaluator import DatasetConfig, SupervisorConfig
from bells_o.supervisors.rest import AzureAnalyzeTextSupervisor
from bells_o import HuggingFaceDataset
from dotenv import load_dotenv

load_dotenv()
```

```python
def bells_target_map(input: str):
    return Result(content_moderation=True if input == "harmful" else False)


dataset_conf: DatasetConfig = DatasetConfig(
    type=HuggingFaceDataset,
    kwargs={
        "name": "brash6/content_moderation_bells_v2",
        "usage": Usage("content_moderation"),
        "target_map_fn": bells_target_map,
        "filters": {"harm_level": ["benign", "harmful"]},
    },
    input_column="question",
    target_column="harm_level",
)
```

```python
supervisor_conf = SupervisorConfig(
    type=AzureAnalyzeTextSupervisor,
    kwargs={"endpoint": "https://content-safety-gratis.cognitiveservices.azure.com/"},
)
```

```python
evaluator = Evaluator(dataset_conf, supervisor_conf, save_dir="../azure_analyze_text", verbose=True)
```

```python
evaluator.run(run_id="run_1", verbose=True, save=True, indices=list(range(100)))
```
