```python
from bells_o.datasets import HuggingFaceDataset
from bells_o.supervisors import HuggingFaceSupervisor
from bells_o import UsageType
from bells_o.preprocessors import RoleWrapper
from tqdm import tqdm
import re
```

```python
usage = UsageType(content_moderation=True, jailbreak=False, prompt_injection=False)
dataset = HuggingFaceDataset("lmsys/toxic-chat", usage, version_name="toxicchat0124")
```

```python
print(dataset[0]["model_output"])
```

```python
def mapping(output):
    return [re.findall(r"<label>(.*)</label>", string)[0] for string in output]


wrapper = RoleWrapper("user", opposite_prompt="\n <think>")

model_kwargs = {"device_map": "auto"}
generation_kwargs = dict(max_new_tokens=512, temperature=0.0000001, do_sample=True)

supervisor = HuggingFaceSupervisor(
    "saillab/x-guard",
    mapping,
    [wrapper],
    apply_chat_template=True,
    model_kwargs=model_kwargs,
    generation_kwargs=generation_kwargs,
)
```

```python
pre_processed = supervisor.pre_process(dataset[1]["model_output"])
output = supervisor.judge(pre_processed)
```
